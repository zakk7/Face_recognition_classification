```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import urllib
import requests

def search_and_download(number_of_images=200):
    x=input ("i want to search for this term:")
    driver = webdriver.Chrome()
    driver.get("https://www.google.com/imghp?hl=fr&authuser=0&ogbl")
    time.sleep(1)
    driver.maximize_window()
#assert "drive" in driver.title
    elem = driver.find_element(By.XPATH, "//textarea[@class='gLFyf']")
#to clear the element if it s already containing text
#time.sleep(4)
    elem.clear()
    elem.send_keys(x)
    elem.send_keys(Keys.ENTER)
#scroll down 2 times and find images
    scroll_count=10
    for _ in range(scroll_count):
         
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        try:
            load_more=driver.find_element(By.CLASS_NAME,"LZ4I").click()
        except:
            
            
            
        

            # Wait for a short duration to allow content to load (optional)
             #time.sleep(2) 
            img_src=[]
            downloaded_img=0
            try: 
                imgs=driver.find_elements(By.XPATH, "//img[@class='rg_i Q4LuWd']")
            except: continue
            
            
    for i,img in enumerate(imgs):
        if downloaded_img < number_of_images :

            #click on the image
            try:
                img.click()
                time.sleep(2)
                # when clicking on the image, it apperas that there is two src one starts with http and the other starts whit data, we want to take the fisrt one
                try:
                    src = driver.find_element(By.XPATH, "//img[@class='sFlh5c pT0Scc iPVvYb']").get_attribute("src")
                    print(src)
                except:
                    continue

                driver.back()
                time.sleep(2)
                if src and 'http' in src:
                    try:
                        response = requests.head(src)
                        if 'Content-Length' in response.headers:
                            print("content,check")
                            image_size_bytes = int(response.headers['Content-Length'])
                            if image_size_bytes > 20000:
                                print("size check")
                                img_src.append(src)
                                
                                if urllib.request.urlretrieve(src, f"C:\\Users\\kouni\\Desktop\\img\\{x}\\captcha{i}.png") :
                                    downloaded_img = downloaded_img + 1
                                    print("downloaded_img",downloaded_img)
                                else:
                                    continue
                        else: 
                            
                            continue
                            
                    except: #urllib.error.HTTPError as err :
                        print ("error downloading the image")
                        continue
            except:
                continue
        else:
            break

    driver.quit()
    return("Done")
#img_data = pd.DataFrame(img_src)
#img_data.to_csv('img_src.csv',index=False)
#assert "No results found." not in driver.page_source
```
